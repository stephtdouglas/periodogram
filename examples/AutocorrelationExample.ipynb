# Using the Autocorrelation Function to find a periodic signal


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import logging
logging.basicConfig(level=logging.DEBUG)

# use seaborn for matplotlib settings
#import seaborn; seaborn.set()
```

## Fake Data
The ACF requires evenly-sampled data without gaps

```python
fake_period = 20.0
cadence = 0.02043423 #d 

t = np.arange(1,80,cadence)
y = np.sin(2*np.pi*t/fake_period) + np.random.normal(scale=0.1,size=len(t))
dy = np.random.normal(scale=0.1,size=len(t))
fig, ax = plt.subplots()
ax.errorbar(t, y, dy, fmt='o')
ax.set_xlabel('time (d)')
ax.set_ylabel('brightness')
```
